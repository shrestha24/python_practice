```python
df = df.drop(['Content','Headquaters'],axis=1)
df.head(10)
```

```python
df['Segment_Category'].value_counts()
```

```python
def remove_theme(col):
  no_theme = col.split(" ")
  if no_theme[-1]=='Cafe' :
    menu = no_theme[-2]
  else:
    menu = no_theme[-1]
  return menu  
```

```python
df['Menu'] = df['Segment_Category'].apply(remove_theme)
```
